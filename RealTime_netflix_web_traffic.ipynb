```python
from flask import Flask, jsonify
from datetime import datetime
import random
import time
from threading import Thread

app = Flask(__name__)

history = []

@app.route("/")
def home():
    return "Flask server running from Jupyter"

@app.route("/metrics_history")
def metrics_history():
    return jsonify(history)

def generate_data():
    while True:
        record = {
            "website": "netflix.com",
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "requests_per_minute": random.randint(30, 60),
            "z_score": round(random.uniform(-1, 1), 2),
            "anomaly": False
        }

        # simulate anomaly
        if random.random() < 0.1:
            record["requests_per_minute"] = random.randint(200, 400)
            record["z_score"] = round(random.uniform(3, 4), 2)
            record["anomaly"] = True

        history.append(record)
        history[:] = history[-60:]

        time.sleep(5)

def run_flask():
    app.run(port=5000, debug=False, use_reloader=False)

Thread(target=generate_data, daemon=True).start()
Thread(target=run_flask, daemon=True).start()

print("✅ Flask running in Jupyter")
```
